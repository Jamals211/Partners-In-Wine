# Red Wine 

```python
#import JSON data with the API
req_red_USA = urllib.request.Request('https://api.globalwinescore.com/globalwinescores/latest/?color=red&limit=5000&country=Usa', data=None, headers=headers, origin_req_host=None, unverifiable=False, method=None)
res_red_USA = urllib.request.urlopen(req_red_USA).read().decode()
obj_red_USA = json.loads(res_red_USA)
#read json into DF
red_USA_df = pd.DataFrame.from_dict(obj_red_USA['results'], orient='columns')
# Delete unnecessary columns
red_USA_df.drop([ 'wine_slug', 'appellation_slug',  'date',  'lwin', 'lwin_11'],axis=1, inplace=True)
#convert vintage to date type
red_USA_df['vintage']= pd.to_datetime(red_USA_df['vintage']).dt.year
red_USA_df['regions']=red_USA_df['regions'].str.get(0)
red_USA_df
```

# White Wine

```python
#import JSON data with the API
req_white_USA = urllib.request.Request('https://api.globalwinescore.com/globalwinescores/latest/?color=white&limit=5000&country=Usa', data=None, headers=headers, origin_req_host=None, unverifiable=False, method=None)
res_white_USA = urllib.request.urlopen(req_white_USA).read().decode()
obj_white_USA = json.loads(res_white_USA)
#read json into DF
white_USA_df = pd.DataFrame.from_dict(obj_white_USA['results'], orient='columns')
# Delete unnecessary columns
white_USA_df.drop([ 'wine_slug', 'appellation_slug',  'date',  'lwin', 'lwin_11'],axis=1, inplace=True)
#convert vintage to date type
white_USA_df['vintage']= pd.to_datetime(white_USA_df['vintage']).dt.year
white_USA_df['regions']=white_USA_df['regions'].str.get(0)
white_USA_df
```

# Min/Max/Avg Temperatures

```python
napa_min_max=napa_min_max.set_index('date')
walla_min_max=walla_min_max.set_index('date')
columbia_min_max=columbia_min_max.set_index('date')
sonoma_min_max=sonoma_min_max.set_index('date')
santa_min_max=santa_min_max.set_index('date')
yakima_min_max=yakima_min_max.set_index('date')
dundee_min_max=dundee_min_max.set_index('date')
willamette_min_max=willamette_min_max.set_index('date')
```

```python
merged_temp=pd.merge(napa_min_max, columbia_min_max, left_index=True, right_index=True)
merged_temp=pd.merge(merged_temp, walla_min_max, left_index=True, right_index=True)
merged_temp=pd.merge(merged_temp, sonoma_min_max, left_index=True, right_index=True)
merged_temp=pd.merge(merged_temp, santa_min_max, left_index=True, right_index=True)
merged_temp=pd.merge(merged_temp, yakima_min_max, left_index=True, right_index=True)
merged_temp=pd.merge(merged_temp, dundee_min_max, left_index=True, right_index=True)
# merged_temp=pd.merge(merged_temp, willamette_min_max, left_index=True, right_index=True)
```

```python
merged_temp.reset_index(inplace=True)
merged_temp['date']=pd.to_datetime(merged_temp['date'])
merged_temp = merged_temp.set_index('date').resample('M').mean()
```

```python
merged_temp.reset_index(inplace= True )
merged_temp['date']=pd.to_datetime(merged_temp['date'])
merged_temp.reset_index(inplace=True)
merged_temp['day'] = merged_temp['date'].dt.day
merged_temp['month'] = merged_temp['date'].dt.month
merged_temp['year'] = merged_temp['date'].dt.year

list_months=[]
for x in merged_temp['month']:
    list_months.append(calendar.month_name[x])
merged_temp['months']=list_months
merged_temp.drop(['date', 'day','month'], axis=1, inplace=True)
```

```python
merged_final_temp=pd.DataFrame()
merged_final_temp['avgTempNapa']=(merged_temp['avgMaxTempNapa']+merged_temp['avgMinTempNapa'])/2
merged_final_temp['avgTempColumbia']=(merged_temp['avgMaxTempColumbia']+merged_temp['avgMinTempColumbia'])/2
merged_final_temp['avgTempWalla']=(merged_temp['avgMaxTempWalla']+merged_temp['avgMaxTempWalla'])/2
merged_final_temp['avgTempSonoma']=(merged_temp['avgMaxTempSonoma']+merged_temp['avgMaxTempSonoma'])/2

merged_final_temp['avgTempSanta']=(merged_temp['avgMaxTempSanta']+merged_temp['avgMinTempSanta'])/2
merged_final_temp['avgTempYakima']=(merged_temp['avgMaxTempYakima']+merged_temp['avgMinTempYakima'])/2
merged_final_temp['avgTempDundee']=(merged_temp['avgMaxTempDundee']+merged_temp['avgMinTempDundee'])/2
merged_final_temp['Year']=merged_temp['year']
merged_final_temp['Months']=merged_temp['months']
merged_final_temp
```
